##### Supervised classification
- Naieve Bayes: Bayes' theorem with "Naive" assumption of conditional independance between every xi
- heavily utilized in autonomous cars
- given a bunch of KNOWN examples, learn a type of traight
- examples consist of labeleled / apriori known "features", and a "label" like good/bad
- if u pick USEFUL features, on a plot, they will be spacially seperated ie: one big grouping that's good, and a seperate grouping that's bad

- the objective of the machine learning process is to define a "decision surface"... the separation between good / bad label for example


##### Intro to naive bayes
- a "decision surface" in the form.... of a line dun dun dun...
- decision surface also known as decision boundry or DS

from https://scikit-learn.org/stable/modules/naive_bayes.html

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)
classifier_gnb = GaussianNB()
y_pred = classifier_gnb.fit(X_train, y_train).predict(X_test)
print("Number of mislabeled points out of a total %d points : %d"
      % (X_test.shape[0], (y_test != y_pred).sum()))
```

#### Basic steps; setup a classifier... (naieve bayes, gaussian naive bayes ect..)
- clf.fit(X,Y)
- clf: the classifier, ie: the machine
- X: features (could be a set of pairs of x,y's... [[-1,2],[3,4]..])
- Y: labels (ie: good/bad, blue/red, 1/2 ect)
